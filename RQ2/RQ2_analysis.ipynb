```python
import pandas as pd
```

## Task 1

```python
#load the commit info dataset
commit_info=pd.read_csv("data/commit-infos.csv",sep="$")
#load the refs dataset
```

```python
commit_time= commit_info.drop(columns=['is-merge', 'authored-date',
       'author-name', 'author-email', 'committer-name', 'committer-email',
       'key', 'commit-msg'])
```

```python
commit_time
```

```python
projs= list(set(commit_time["repo"].tolist()))
```

```python
cts=[] #hold the list of dataframes separated by project
for proj in projs:
    df= commit_time[commit_time["repo"]==proj]
    df.reset_index(inplace=True)
    cts.append(df)
```

```python
#relative time= (current_index-first_index)/(total_commits) or since the index starts from 0 index*100/len(data_frame)
def compute_relative_time(row,total):
    return (row.name*100)/total
```

```python
#apply for all dataframes
for c in cts:
    total=len(c.index)
    c["relative_time"]=c.apply(compute_relative_time, args=(total,), axis=1)
```

```python
#now re-merge the dataframes and finish Task 1
commit_time_final=pd.concat(cts)
```

```python
commit_time_final.columns
```

## Task 2 : merge ther refactorings dataset with commit time dataset

```python
#Load the refactorings data
refs= pd.read_csv("data/refactorings.csv", sep="$")
refs
```

```python
refs.dtypes
```

```python
commit_time_final.dtypes
```

```python
#Join the refs and commit time
ref_time=refs.merge(commit_time_final, left_on='sha', right_on='sha')
```

```python
ref_time= ref_time.drop(columns=['key', 'ref_desc',
       'method_lines', 'lines', 'index', 'repo_y',])
```

```python
ref_time.columns=['repo', 'sha', 'ref_id', 'ref_type', 'file_name', 'is_dac',
       'commit-date', 'relative_time']
```

```python
ref_time=ref_time.drop(columns=['sha', 'file_name'])
```

```python
def rename_is_dac(is_dac):
    if is_dac==True:
        return "DAC"
    else:
        return "Regular"
ref_time["is_dac"]=ref_time["is_dac"].apply(rename_is_dac)
```

```python
ref_time.to_csv("data/RQ2-dataset.csv", index=False)
```

## Task4: For DAC refactorings, get the time distribution grouping by refactoring types, for the top 10 most prevalent refactoring

```python
dac_ref_time=ref_time[ref_time["is_dac"]=="DAC"]
dac_ref_time
```

```python
df=dac_ref_time.groupby(by="ref_type")['relative_time'].describe().reset_index()
#df["ref_type"]=df.index
```

```python
dac_ref_time_per_type=df.sort_values('count', ascending=False).round(3).head(10)
dac_ref_time_per_type.round({'count':0})
dac_ref_time_per_type['count']=dac_ref_time_per_type['count'].astype(int)
dac_ref_time_per_type=dac_ref_time_per_type.sort_values('50%', ascending=False)
dac_ref_time_per_type
```

```python
dac_ref_time_per_type.columns
```

```python
#get a latex table from dac_ref_time_per_type
dac_ref_time_per_type_t10_latex=dac_ref_time_per_type.to_latex(columns=['ref_type', 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'], index=False)
print(dac_ref_time_per_type_t10_latex)
```
